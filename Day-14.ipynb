## Setup

Use the `input` file if present, otherwise use the sample input.

```python
import numpy as np

def get_input():
  try:
    return open("input", "r").read().splitlines()
  except FileNotFoundError:
    input = """O....#....
O.OO#....#
.....##...
OO.#O....O
.O.....O#.
O.#..O.#.#
..O..#O..O
.......O..
#....###..
#OO..#...."""
    return input.split("\n")

input = get_input()
input = [[c for c in cc] for cc in input]
nin = np.array(input)
print(nin)
print(nin.shape)

```

```python
def roll_rocks(grid):
    for i in range(grid.shape[0]-2, -1, -1):
        for j in range(grid.shape[1]):
            if grid[i, j] == "O":
                for ii in range(i+1, grid.shape[0]+1):
                    if ii == grid.shape[0] or grid[ii, j] != ".":
                        grid[i, j] = "."
                        grid[ii-1, j] = "O"
                        break
```

```python
def get_load(grid):
    res = 0
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            if grid[i, j] == "O":
                res += i + 1
    return res
```

## Part 1

```python
res = 0

p1 = np.copy(nin)
p1 = np.flip(p1, axis=0)
print(p1)

roll_rocks(p1)

print()
print(p1)
for i in range(p1.shape[0]):
    for j in range(p1.shape[1]):
        if p1[i, j] == "O":
            res += i + 1

res = get_load(p1)
print(res)
assert(res == 136)
```

## Part 2

```python
res_arr = []

p2 = np.copy(nin)
p2 = np.flip(p2, axis=0)

for i in range(1000):
    roll_rocks(p2)
    p2 = np.rot90(p2)
    roll_rocks(p2)
    p2 = np.rot90(p2)
    roll_rocks(p2)
    p2 = np.rot90(p2)
    roll_rocks(p2)
    p2 = np.rot90(p2)
    load = get_load(p2)
    res_arr.append(load)

# Look for something resembling a repeating cycle, start at end
print(res_arr)
seq = []
seqloc = None
for i in range(len(res_arr)):
    seq = res_arr[-i:]
    print(seq, res_arr[-2*i:-i])
    if seq == res_arr[-2*i:-i]:
        print("found seq", seq)
        seqloc = i
        break
```
